```python
import Ipynb_importer
from PIL import Image
import json
import os
import numpy as np
from model import AlexNet_v1, AlexNet_v2
import matplotlib.pyplot as plt
%matplotlib inline

os.environ["CUDA_VISIBLE_DEVICES"] = "-1" # 指定使用CPU
```

```python
def process_data(img_path):
    assert os.path.exists(img_path), "file: {} dose not exist.".format(img_path)
    im_height = 224
    im_width = 224
    
    # 加载图片
    with Image.open(img_path) as img:
        # resize
        img = img.resize((im_width, im_height))
        plt.imshow(img)
        # 归一化
        img = np.array(img) / 225.0
        # 拓展batch维度
        img = np.expand_dims(img, 0)
    return img
```

```python
# 加载图片
img_path = "./data/rose_demo.jpg"
img = process_data(img_path)
```

```python
# 获取lable
try:
    json_path = "./class_indices.json"
    assert os.path.exists(json_path), "file: {} does not exist.".format(json_path)
    with open(json_path) as json_file:
        class_indices = json.load(json_file)
        print(class_indices)
except Exception as e:
    print(e)
    exit(-1)
```

```python
# 实例化模型
model = AlexNet_v1(num_classes=5)
# 加载模型参数
model_weight_path = "./save_weights/myAlexNet_cpu.h5"
assert os.path.exists(model_weight_path), "file: {} does not exist.".format(model_weight_path)
model.load_weights(model_weight_path)
```

```python
# 开始预测
output = np.squeeze(model.predict(img)) #压缩掉Batch维度，返回概率分布
predict_class_index = np.argmax(output) #获取概率值最大的index

print(f"[预测类别]:{class_indices[str(predict_class_index)]}\t[置信度]:{(output[predict_class_index]):.3}")
```
